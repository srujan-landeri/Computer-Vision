```python
import tensorflow as tf
import warnings

warnings.filterwarnings('ignore')
```

```python
tensor = tf.constant([1,2,3,4,-1,-3,-4,-6])
print("abs:\n",tf.abs(tensor))

tensor = tf.constant([1,4,9,16], dtype=tf.dtypes.float32)
print("sqrt:\n",tf.sqrt(tensor))

x1 = tf.constant([1,2,3,4])
x2 = tf.constant([1,2,3,4])

print("add:\n",tf.add(x1,x2))
print("multiply:\n",tf.multiply(x1,x2))
print("subtract:\n",tf.subtract(x1,x2))
print("divide:\n",tf.divide(x1,x2))
```

```python
x1 = tf.constant([[5,3,6,6,4,6],], dtype=tf.float32)
x2 = tf.constant([[7],[5],[3]], dtype=tf.float32)

print(tf.math.multiply(x1,x2))
```

```python
# position of the max element is given
x1 = tf.constant([112,20,30,3,6])
print(tf.math.argmax(x1))

x2  = tf.constant(
    [
        [1,2,3,4],
        [7,8,9,0],
        [5,6,7,8]
    ]
)

print(tf.math.argmax(x2),1) # returns the ind of max element for each column
```

```python
print(tf.math.reduce_max(x2,axis=1)) # reducing on basis of column values
print(tf.math.reduce_max(x2,axis=0)) # reducing on basis of row values

# other methods .... 
# tf.math.reduce_mean
# tf.math.reduce_min
# tf.math.reduce_sum
# tf.math.reduce_max 
# tf.math.reduce_variance 
# tf.math.reduce_prod
```
